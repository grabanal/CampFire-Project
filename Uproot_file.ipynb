```python
import uproot
import numpy as np
import matplotlib.pyplot as plt

path = "/Users/grabanal/Documents/Harvard/2019_II_Summer_CURRENT/CampFire-Project/"
s_filename = path + "signal_output.root"
b_filename = path + "BG_output.root"
quantities = {"PtLGammaNu":[r"$P_T^{\ell+\gamma+MET}$",r"$\frac{\mathrm{GeV}}{c}$"],
              "MET"       :[r"Missing E$_\mathrm{T}$","GeV"],
              "HT"        :[r"$H_T$","GeV"],
              "Plq"       :[r"$P_\ell\cdot q$",r"${ \left(\frac{\mathrm{GeV}}{c}\right) }^2$"]}
for q in quantities.keys():
    # importing the TH1F's using uproot
    s_cts = uproot.open(s_filename)[q].numpy()[0] # an array of contents
    s_bin = uproot.open(s_filename)[q].numpy()[1] # an array of bins
    b_cts = uproot.open(b_filename)[q].numpy()[0] # an array of contents
    b_bin = uproot.open(b_filename)[q].numpy()[1] # an array of bins
    # playing with the arrays
    fig = plt.figure()
    plt.title(quantities[q][0],  fontsize=20)
    plt.xlabel(quantities[q][1], fontsize=18)
    plt.ylabel("Event fraction / 10 GeV", fontsize=12)
    plt.plot(s_bin[:-1],s_cts/s_cts.sum(),drawstyle='steps-pre',label=r"H$^\pm$H$^{\mp\mp}$")
    plt.plot(b_bin[:-1],b_cts/b_cts.sum(),drawstyle='steps-pre',label=r"$W\,\gamma$")
    plt.legend(prop={'size': 18})
    plt.show()
    fig.savefig("figure" + q + ".pdf",bbox_inches="tight")
```
